# Công cụ Dự đoán Khách hàng Rời bỏ (Customer Churn Prediction Demo)

### Mô tả chung
Ứng dụng web demo giúp bạn trực quan hiểu cách mô hình Machine Learning dự đoán khả năng một khách hàng viễn thông sẽ rời bỏ dịch vụ (Churn) trong 30-60 ngày tới.

Mô hình được huấn luyện trên bộ dữ liệu Customer Churn với hơn 7.000 khách hàng thực tế. Độ chính xác trung bình 80-85% (Random Forest / XGBoost).

### Ý nghĩa của từng tham số bạn có thể thay đổi

| Tham số                       | Ý nghĩa thực tế                                                                                 | Tác động đến nguy cơ rời bỏ (Churn)                                                                 |
|-------------------------------|-------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------|
| Tenure (Thâm niên)            | Số tháng khách hàng đã dùng dịch vụ liên tục                                                    | Càng cao → càng trung thành. Khách mới (1-3 tháng) có nguy cơ cao nhất                              |
| Monthly Charges               | Cước phí trung bình mỗi tháng                                                                   | Cao → rủi ro tăng (nhạy cảm giá), nhưng nếu kèm nhiều dịch vụ cao cấp thì lại giảm rủi ro            |
| Contract                      | Loại hợp đồng                                                                                   | Theo tháng → NGUY HIỂM NHẤT<br>1 năm → trung bình<br>2 năm → an toàn nhất                            |
| Internet Service              | Loại đường truyền                                                                               | No → rất trung thành<br>DSL → ổn định<br>Fiber optic → rủi ro cao hơn (cạnh tranh gay gắt)           |
| 
| Payment Method                | Phương thức thanh toán                                                                          | Electronic check → rủi ro cao nhất<br>Bank transfer / Credit card (automatic) → rủi ro thấp nhất    |
| Tech Support / Online Security / Online Backup / Device Protection | Các dịch vụ phụ trợ                                                                    | Có đăng ký → khách “dính” nhiều → ít rời bỏ hơn                                                      |
| Senior Citizen                | Người cao tuổi (65+)                                                                            | Thường trung thành hơn                                                                              |
| Partner / Dependents          | Có vợ/chồng hoặc người phụ thuộc                                                                | Có gia đình → xu hướng ở lại cao hơn                                                                |

### Cơ chế hoạt động (đơn giản hóa)

Mô hình tính điểm rủi ro theo công thức có trọng số:
Churn Score = w₁×Tenure + w₂×Contract + w₃×MonthlyCharges + w₄×PaymentMethod + …

Các trọng số quan trọng nhất (thực tế học được):
- Contract = Month-to-month → +2.5 ~ +3.5 điểm (rất nặng)
- Payment = Electronic check → +0.8 ~ +1.2 điểm
- Mỗi +12 tháng Tenure → -1 ~ -1.5 điểm
- Có Tech Support → -0.8 điểm
- Fiber optic → +0.6 ~ +1 điểm

Sau đó chuyển thành xác suất:
- Score > 0.5 → Dự đoán “Rời bỏ”
- Score ≤ 0.5 → Dự đoán “Ở lại”

### Ví dụ thay đổi tham số

| Thay đổi                                      | Kết quả dự kiến                 | Giải thích                                                      |
|-----------------------------------------------|---------------------------------|-----------------------------------------------------------------|
| Tenure 12 → 1 tháng                           | ~65-75% (Đỏ)                    | Khách mới + hợp đồng tháng = nhóm nguy cơ cao nhất              |
| Contract → Two year                           | ~3-8% (Xanh lá)                 | Bị ràng buộc 24 tháng → hầu như không rời bỏ                    |
| Payment → Credit card (automatic)             | Giảm thêm 10-20%                | Thanh toán tự động → ít cơ hội suy nghĩ hủy                      |
| Bật hết Tech Support, Security, Backup…       | Giảm thêm 15-25%                | Dùng nhiều dịch vụ → ngại chuyển nhà cung cấp                   |
